```python
import pandas as pd

proyectos = pd.read_excel("D:\\data\\ProyectosFinalizados.xlsx", sheet_name="base")

def limpiar_columna(columna):
    # 1. Reemplazar caracteres especiales
    columna = columna.replace(" ", "_") \
        .replace("á", "a") \
        .replace("é", "e") \
        .replace("í", "i") \
        .replace("ó", "o") \
        .replace("ú", "u") \
        .replace("ñ", "n") \
        .replace("+", "_") \
        .replace(".", "_") \
        .replace("(", "_") \
        .replace(")", "")
    import re
    # 2. Poner un guion bajo entre cambio de letras de minúscula a mayúsculas
    columna = re.sub("([a-z])([A-Z])", "\\1_\\2", columna)
    # 3. Poner un guion bajo entre ID y letra
    columna = re.sub("ID([A-Za-z])", "ID_\\1", columna)
    # 4. Poner un guion bajo entre número y letra
    columna = re.sub("([0-9])([A-Za-z])", "\\1_\\2", columna)
    # 4. Poner un guion bajo entre letra y número
    columna = re.sub("([A-Za-z])([0-9])", "\\1_\\2", columna)
    # 5. Quitar un guion bajo entre letra y número
    columna = re.sub("([A-Za-z])_([0-9])", "\\1\\2", columna)
    # 6. Poner un guion bajo entre letra y número final
    columna = re.sub("([A-Za-z])([0-9])$", "\\1_\\2", columna)
    # 7. Convertir a mayúsculas
    columna = columna.upper()
    return columna

proyectos.columns = map(limpiar_columna, proyectos.columns.values)

proyectos.head()
```

```python
proyectos["OBSERVACIONES"].sample(10)
```

```python
observaciones_list = proyectos["OBSERVACIONES"].dropna().values

observaciones_list
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()

X = vectorizer.fit_transform(observaciones_list)

X[:5].toarray()
```

```python
vectorizer.vocabulary_
```

```python
from operator import itemgetter

sorted(vectorizer.vocabulary_.items(), key=itemgetter(0), reverse=True)
```

```python
from collections import defaultdict

words_freq = defaultdict(int)

for word, index in list(vectorizer.vocabulary_.items()):
    words_freq[word] = X.toarray()[:, index].sum()

words_freq
```

```python
sorted(words_freq.items(), key=itemgetter(1), reverse=True)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.pipeline import Pipeline

vocabulary = ["líder", "esfuerzo", "desarrollo", "vulnerabilidades", "proyecto"]

pipe = Pipeline([
    ("count", CountVectorizer(vocabulary=vocabulary)),
    ("tfid", TfidfTransformer())
])

pipe.fit(observaciones_list)

pipe["count"].fit_transform(observaciones_list).toarray()
```
