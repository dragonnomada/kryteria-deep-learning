```python
import networkx as nx

G = nx.DiGraph()

G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)

G.add_nodes_from([5, 6, 7])

G.nodes
```

```python
G.add_edge(1, 2)

G.add_edges_from([(2, 4), (3, 4), (6, 1), (5, 7), (3, 7)])

G.edges
```

```python
nx.draw(G, with_labels=True, font_weight='bold')
```

```python
import matplotlib.pyplot as plt

plt.figure()

pos = nx.spring_layout(G)

nx.draw(G, pos, with_labels=True, font_weight="bold", edge_color="black", width=1, linewidths=1,
    node_size=500, node_color="pink", alpha=0.9,
    labels={node: node for node in G.nodes()})

nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels={(1, 2): "A",
                 (2, 4): "B",
                 (6, 1): "C"},
    font_color="red"
)

plt.show()
```

```python
import networkx as nx

G = nx.DiGraph()

G.add_nodes_from([1, 2, 3, 4, 5, 6, 7])
G.add_edges_from([
    (1, 2, { "label": "A", "length": 1 }),
    (2, 4, { "label": "B", "length": 1 }),
    (3, 4, { "label": "C", "length": 2 }),
    (6, 1, { "label": "D", "length": 2 }),
    (5, 7, { "label": "E", "length": 1 }),
    (3, 7, { "label": "F", "length": 1 })
])

import matplotlib.pyplot as plt

plt.figure()

pos = nx.spring_layout(G)

nx.draw(G, pos, with_labels=True, font_weight="bold", edge_color="black", width=1, linewidths=1,
    node_size=500, node_color="pink", alpha=0.9,
    labels={node: node for node in G.nodes()})

nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels= dict([((n1, n2), "{} ({})".format(d["label"], d["length"])) for n1, n2, d in G.edges(data=True)]),
    font_color="red"
)

plt.show()
```
