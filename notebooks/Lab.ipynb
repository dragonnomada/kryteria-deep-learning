```python
from sklearn.datasets import load_iris

iris = load_iris()

X = iris.data
y = iris.target

X, y, iris.target_names, iris.feature_names
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y)

X_train, X_test, y_train, y_test
```

```python
X.shape, X_train.shape, X_test.shape, X_train.shape[0] / X.shape[0], X_test.shape[0] / X.shape[0]
```

```python
y.shape, y_train.shape, y_test.shape, y_train.shape[0] / y.shape[0], y_test.shape[0] / y.shape[0]
```

```python
y.reshape((150, 1)),y.reshape((-1,))
```

```python
from sklearn.neighbors import KNeighborsClassifier

estimator = KNeighborsClassifier(n_neighbors=3)

estimator.fit(X_train, y_train)

y_predicted = estimator.predict(X_test)

print("Precisión: {:.1f}".format((y_test == y_predicted).mean() * 100))
```

```python
X_test, y_test, y_predicted
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(estimator, X, y, scoring='accuracy')

print(scores)
```

```python
accuracy = scores.mean() * 100

accuracy
```

```python
import pandas as pd

games = pd.read_csv("../data/basketball/NBA_2016_games.csv", parse_dates=["Date"])

games.columns = ["DATE", "START", "VISITOR_TEAM", "VISITOR_POINTS", "HOME_TEAM", "HOME_POINTS", "BOX", "OT", "ATTEND", "ARENA", "NOTES"]

games
```

```python
games["HOME_WIN"] = games["VISITOR_POINTS"] < games["HOME_POINTS"]

games.sample(5)
```

```python
games["HOME_WIN"].mean()
```

```python
from collections import defaultdict

won_last = defaultdict(int)

for index, row in games.iterrows():

    home_team = row["HOME_TEAM"]
    visitor_team = row["VISITOR_TEAM"]

    games.at[index, "HOME_LAST_WIN"] = bool(won_last[home_team])
    games.at[index, "VISITOR_LAST_WIN"] = bool(won_last[visitor_team])

    if row["HOME_WIN"]:
        # print("WIN HOME: {} vs {}".format(home_team, visitor_team))
        won_last[home_team] = 1
        won_last[visitor_team] = 0
    else:
        # print("WIN VISITOR: {} vs {}".format(home_team, visitor_team))
        won_last[home_team] = 0
        won_last[visitor_team] = 1

games
```

```python
X = games[ ["HOME_LAST_WIN", "VISITOR_LAST_WIN"] ].values.astype(int)
y = games["HOME_WIN"].values.astype(int)

X, y
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

scores = cross_val_score(DecisionTreeClassifier(random_state=14), X, y, scoring='accuracy')

100 * scores.mean()
```

```python
standings = pd.read_csv("../data/basketball/NBA_2015_standings.csv")

standings
```

```python
for index, row in games.iterrows():

    home_team = row["HOME_TEAM"]
    visitor_team = row["VISITOR_TEAM"]

    home_rank = standings[ (standings["Team"] == home_team) ]["Rk"].values[0]
    visitor_rank = standings[ (standings["Team"] == visitor_team) ]["Rk"].values[0]

    games.at[index, "HOME_RANK"] = home_rank
    games.at[index, "VISITOR_RANK"] = visitor_rank
    games.at[index, "HOME_RANK_HIGHER"] = home_rank < visitor_rank

games
```

```python
X = games[ ["HOME_LAST_WIN", "VISITOR_LAST_WIN", "HOME_RANK_HIGHER"] ].values.astype(int)
y = games["HOME_WIN"].values.astype(int)

X, y
```

```python
scores = cross_val_score(DecisionTreeClassifier(random_state=14), X, y, scoring='accuracy')

100 * scores.mean()
```

```python
last_winner_match_teams =defaultdict(str)

for index, row in games.iterrows():

    home_team = row["HOME_TEAM"]
    visitor_team = row["VISITOR_TEAM"]

    match_team = tuple(sorted([home_team, visitor_team]))

    home_win_last_match = False

    if last_winner_match_teams[match_team] == home_team:
        home_win_last_match = True

    games.at[index, "HOME_WIN_LAST_MATCH"] = home_win_last_match

    winner = visitor_team

    if row["HOME_WIN"]:
        winner = home_team

    last_winner_match_teams[match_team] = winner

games
```

```python
X = games[ ["HOME_LAST_WIN", "VISITOR_LAST_WIN", "HOME_RANK_HIGHER", "HOME_WIN_LAST_MATCH"] ].values.astype(int)
y = games["HOME_WIN"].values.astype(int)

X, y
```

```python
scores = cross_val_score(DecisionTreeClassifier(random_state=14), X, y, scoring='accuracy')

100 * scores.mean()
```

```python
from sklearn.ensemble import RandomForestClassifier

scores = cross_val_score(RandomForestClassifier(random_state=14), X, y, scoring='accuracy')

100 * scores.mean()
```

```python
from sklearn.model_selection import GridSearchCV

parameter_space = {
    "max_features": [2, 10],
    "n_estimators": [100, 200],
    "criterion": ["gini", "entropy"],
    "min_samples_leaf": [2, 4, 6]
}

estimator = RandomForestClassifier(random_state=14)

grid = GridSearchCV(estimator, parameter_space)

scores = cross_val_score(grid, X, y, scoring='accuracy')

100 * scores.mean()
```

```python
import pandas as pd

ratings = pd.read_csv("../data/ml-latest-small/ratings.csv")

ratings.sample(5)
```
