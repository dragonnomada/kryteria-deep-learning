```python
from sklearn.datasets import load_iris

iris = load_iris()

X = iris.data
y = iris.target

X, y, iris.target_names, iris.feature_names
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y)

X_train, X_test, y_train, y_test
```

```python
X.shape, X_train.shape, X_test.shape, X_train.shape[0] / X.shape[0], X_test.shape[0] / X.shape[0]
```

```python
y.shape, y_train.shape, y_test.shape, y_train.shape[0] / y.shape[0], y_test.shape[0] / y.shape[0]
```

```python
y.reshape((150, 1)),y.reshape((-1,))
```

```python
from sklearn.neighbors import KNeighborsClassifier

estimator = KNeighborsClassifier(n_neighbors=3)

estimator.fit(X_train, y_train)

y_predicted = estimator.predict(X_test)

print("Precisión: {:.1f}".format((y_test == y_predicted).mean() * 100))
```

```python
X_test, y_test, y_predicted
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(estimator, X, y, scoring='accuracy')

print(scores)
```

```python
accuracy = scores.mean() * 100

accuracy
```

```python
X_broken = np.array(X)

X_broken[:,::3] /= 4

X_broken
```

```python
estimator = KNeighborsClassifier()

scores = cross_val_score(estimator, X_broken, y, scoring='accuracy')

scores.mean()
```

```python
from sklearn.pipeline import Pipeline

sequence = Pipeline([
    ('scale', MinMaxScaler()),
    ('estimate', KNeighborsClassifier())
])

scores = cross_val_score(sequence, X_broken, y, scoring='accuracy')

scores.mean()
```
