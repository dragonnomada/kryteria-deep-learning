### scipy - Herramientas científicas

````python
from scipy import linalg
import numpy as np

# x + 2y = 19
# 2x + 3y = 31

A = np.array([
    [1, 2],
    [2, 3],
])

b = np.array([19, 31])

x = linalg.solve(A, b)

A, b, x
````

````python
from scipy.optimize import minimize
import numpy as np

x0 = np.array([1, 1])

def f(x):
    return (x[0] - 5) ** 2 + (x[1] - 7) ** 2

result = minimize(f, x0, method="nelder-mead")

result
````

### Pillow - Manipulación de imágenes

````python
from PIL import Image

im = Image.open("../data/coca-cola.png")

im.resize((80, 80))
````

````python
im.resize((80, 80)).crop((20, 20, 60, 60))
````

````python
im.resize((80, 80)).rotate(45).crop((20, 20, 60, 60))
````

````python
from PIL import ImageFilter

im.resize((80, 80)).rotate(45).crop((20, 20, 60, 60)).filter(ImageFilter.SMOOTH)
````

### scikit-learn - Herramietas de machine learning

````python
from sklearn.svm import SVC

X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([
    1,
    0,
    0,
    1
])

clf = SVC()

clf.fit(X, y)

clf.predict([
    [0, 0],
    [0, 0.5],
    [0, 1],
    [0.5, 0.5],
    [0.5, 1],
    [1, 0],
    [1, 0.5],
    [1, 1],
])
````
